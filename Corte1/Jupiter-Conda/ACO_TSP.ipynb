```python
from ipycanvas import Canvas
from PIL import Image
import numpy as np
import os
```

```python
# Crear lienzo interactivo
canvas = Canvas(width=200, height=200, background_color="white")

# Función para capturar el dibujo como una imagen
def get_drawing():
    img = Image.fromarray(canvas.to_image_data()).convert("L")  # Escala de grises
    img = img.resize((28, 28))  # Redimensionar a 28x28 píxeles
    return np.array(img)

# Función para guardar el dibujo
def save_drawing(class_name, count):
    os.makedirs(f"data/{class_name}", exist_ok=True)
    img = get_drawing()
    filepath = f"data/{class_name}/{count}.png"
    Image.fromarray(img).save(filepath)
    print(f"Dibujo guardado: {filepath}")

# Mostrar el lienzo
display(canvas)
```
